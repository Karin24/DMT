```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('training_set_VU_DM.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.astype({"date_time": 'datetime64[ns]'})
df.info()
```

```python
df['price_per_room'] = df['price_usd'] / df['srch_room_count']/df['srch_length_of_stay']
#df['price_per_room'].value_counts().to_frame().style.bar()"
```

```python
df['total_people'] = df['srch_adults_count']+df['srch_children_count']
df['people_per_room'] = df['total_people']/ df['srch_room_count']
df['price_per_person'] = df['price_usd']/df['total_people']
```

```python
df = df.fillna(0)
df = df.replace(to_replace=-1, value=0)
```

```python
df['prop_location_score'] = df['prop_location_score1']+df['prop_location_score2']
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
comp_rate = pca.fit_transform(df[['comp{0}_rate'.format(i + 1) for i in range(8)]].fillna(0))
comp_inv = pca.fit_transform(df[['comp{0}_inv'.format(i + 1) for i in range(8)]].fillna(0))
comp_rate_percent_diff = pca.fit_transform(df[['comp{0}_rate_percent_diff'.format(i + 1) for i in range(8)]].fillna(0))
```

```python
comp_rate = pd.DataFrame(data=comp_rate)
comp_inv = pd.DataFrame(data=comp_inv)
comp_rate_percent_diff = pd.DataFrame(data=comp_rate_percent_diff)
```

```python
f = {"comp_rate":comp_rate[0],"comp_inv": comp_inv[0], "comp_rate_percent_diff":comp_rate_percent_diff[0]}
df_comp = pd.DataFrame(data = f)
df_comp
```

```python
df = pd.concat([df,df_comp], axis=1)
df.head()
```

```python
x_train = df[['srch_id','visitor_location_country_id','prop_country_id','prop_brand_bool', 'prop_starrating', 'prop_review_score','srch_children_count','total_people','people_per_room','price_per_person', 'srch_length_of_stay', 'srch_query_affinity_score', 'prop_location_score', 'comp_rate', 'comp_inv', 'comp_rate_percent_diff','price_per_room','promotion_flag', 'orig_destination_distance']]
y_train = df[['srch_id','prop_id']]
```

```python
test = pd.read_csv('test_set_VU_DM.csv')
test.head()
```

```python
test.isnull().sum()
```

```python
test = test.fillna(0)
test = test.replace(to_replace=-1, value=0)
```

```python
comp_rate = pd.DataFrame(data=comp_rate)
comp_inv = pd.DataFrame(data=comp_inv)
comp_rate_percent_diff = pd.DataFrame(data=comp_rate_percent_diff)
```

```python
df_comp_test = {"comp_rate":comp_rate[0],"comp_inv":comp_inv[0],"comp_rate_percent_diff":comp_rate_percent_diff[0]}
df_comp_test = pd.DataFrame(data=df_comp_test)
df_comp_test.head()
```

```python
test = pd.concat([test,df_comp_test], axis=1)
test.head()
```
